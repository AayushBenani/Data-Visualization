```python
import pandas as pd
from bokeh .io import show, output_file
from bokeh.plotting import figure
from bokeh.layouts import row, column
from operator import itemgetter
import math
def analyze_attribute_distribution_v2(attribute_idx):
  distribution_counts = {}
  data_source = open('/content/adult.data', "r")
  for data_row in data_source:
    row_elements = data_row.rstrip().split(",")
    if len(row_elements) > attribute_idx:
        attribute_val = row_elements[attribute_idx].strip()
        if attribute_val != '?':
            if attribute_val in distribution_counts:
              distribution_counts[attribute_val]+=1
            else:
              distribution_counts[attribute_val]=1
  data_source.close()
  sorted_attribute_keys =[]
  sorted_attribute_values = []
  for (k_val,v_count) in sorted(distribution_counts.items(), key = itemgetter(1), reverse = True):
    sorted_attribute_keys.append(k_val)
    sorted_attribute_values.append(distribution_counts[k_val])
  return sorted_attribute_keys,sorted_attribute_values


file_handle = open('/content/adult.data', "r")
age_distribution_data = {}
for data_entry in file_handle:
  data_entry = data_entry.strip()
  if not data_entry:
      continue
  entry_parts = data_entry.split(",")
  if entry_parts[0].strip() != '?':
      original_age_value = int(entry_parts[0])
      if original_age_value < 20:
        age_group_identifier = 10
      else:
        age_group_identifier  = int(original_age_value/10)*10
      if age_group_identifier in age_distribution_data:
        age_distribution_data[age_group_identifier]+=1
      else:
        age_distribution_data[age_group_identifier]=1
print(age_distribution_data)
file_handle.close()
age_group_labels_list = []
age_group_counts_list = []
for age_bin_value in range(10,110,10):
  if age_bin_value in age_distribution_data:
    if age_bin_value == 10:
      age_group_labels_list.append("<20")
    else:
      age_group_labels_list.append(str(age_bin_value)+ "\'s")
    age_group_counts_list.append(age_distribution_data[age_bin_value])
output_file("SIMPLE_DASHBOARD.html")
age_figure_plot = figure(x_range = age_group_labels_list, height = 400, title = "Age Group Distribution")
age_figure_plot.vbar(x = age_group_labels_list, top = age_group_counts_list, width = 0.5, color = "black", alpha = 0.5)
age_figure_plot.xgrid.grid_line_color = None
age_figure_plot.ygrid.grid_line_color = None
age_figure_plot.xaxis.major_tick_line_color = None
age_figure_plot.yaxis.major_tick_line_color = None
age_figure_plot.xaxis.minor_tick_line_color = None
age_figure_plot.yaxis.minor_tick_line_color = None
age_figure_plot.outline_line_color = None
(education_category_keys, education_category_values) = analyze_attribute_distribution_v2(3)
education_figure_plot = figure(x_range = education_category_keys, height = 400, title = "Education Distribution")
education_figure_plot.vbar(x = education_category_keys, top = education_category_values, width = 0.5, color = "black", alpha = 0.5)
education_figure_plot.xaxis.major_label_orientation = math.pi/2
education_figure_plot.xgrid.grid_line_color = None
education_figure_plot.ygrid.grid_line_color = None
education_figure_plot.xaxis.major_tick_line_color = None
education_figure_plot.yaxis.major_tick_line_color = None
education_figure_plot.xaxis.minor_tick_line_color = None
education_figure_plot.yaxis.minor_tick_line_color = None
education_figure_plot.outline_line_color = None
(marital_status_keys, marital_status_values) = analyze_attribute_distribution_v2(5)
marital_status_figure_plot = figure(x_range = marital_status_keys, height = 400, title = "Marital Status Distribution")
marital_status_figure_plot.vbar(x = marital_status_keys, top = marital_status_values, width = 0.5, color = "black", alpha = 0.5)
marital_status_figure_plot.xaxis.major_label_orientation = math.pi/2
marital_status_figure_plot.xgrid.grid_line_color = None
marital_status_figure_plot.ygrid.grid_line_color = None
marital_status_figure_plot.xaxis.major_tick_line_color = None
marital_status_figure_plot.yaxis.major_tick_line_color = None
marital_status_figure_plot.xaxis.minor_tick_line_color = None
marital_status_figure_plot.yaxis.minor_tick_line_color = None
marital_status_figure_plot.outline_line_color = None
(working_class_keys, working_class_values) = analyze_attribute_distribution_v2(1)
working_class_figure_plot = figure(x_range = working_class_keys, height = 400, title = "Working Class Distribution")
working_class_figure_plot.vbar(x = working_class_keys, top = working_class_values, width = 0.5, color = "black", alpha = 0.5)
working_class_figure_plot.xaxis.major_label_orientation = math.pi/2
working_class_figure_plot.xgrid.grid_line_color = None
working_class_figure_plot.ygrid.grid_line_color = None
working_class_figure_plot.xaxis.major_tick_line_color = None
working_class_figure_plot.yaxis.major_tick_line_color = None
working_class_figure_plot.xaxis.minor_tick_line_color = None
working_class_figure_plot.yaxis.minor_tick_line_color = None
working_class_figure_plot.outline_line_color = None
show(row(column(age_figure_plot,education_figure_plot),column(marital_status_figure_plot,working_class_figure_plot)))
```
