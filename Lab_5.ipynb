```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import LabelSet, ColumnDataSource
incm_cnts = {}

with open('adult.data', 'r') as file:
    for dash in file:
        incm = dash.strip().split(',')[-1].strip()
        if incm:
            incm_cnts[incm] = incm_cnts.get(incm, 0) + 1
w = []
f = []
for kay, vls in incm_cnts.items():
    w.append(kay)
    f.append(vls)
source = ColumnDataSource(data=dict(x=w, y=f))
ot = figure(x_range=w, height=600, width=600, title="Income Distribution")
ot.vbar(x='x', top='y', width=0.5, source=source, color='steelblue')
labels = LabelSet(x='x', y='y', text='y', y_offset=5, source=source)
ot.add_layout(labels)
ot.yaxis.visible = False
ot.xaxis.minor_tick_line_color = None
ot.xgrid.visible = False
ot.ygrid.visible = False
ot.outline_line_color = None

output_notebook()
show(ot)
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import LabelSet, ColumnDataSource
exp = ['R&D Spend', 'Administration', 'Marketing Spend']
acc = {'R&D Spend': 0, 'Administration': 0, 'Marketing Spend': 0}
cnt = 0

with open('/content/50_Startups.csv', 'r') as file:
    next(file)
    for dash in file:
        prts = dash.strip().split(',')
        acc['R&D Spend'] += float(prts[0])
        acc['Administration'] += float(prts[1])
        acc['Marketing Spend'] += float(prts[2])
        cnt += 1
w = exp
f = [round(acc[key] / cnt, 2) for key in exp]
source = ColumnDataSource(data=dict(x=w, y=f))
ot = figure(x_range=w, width=700, height=600,
           title="Average Startup Spending by Category")
ot.min_border_left = 80
ot.min_border_right = 80
ot.min_border_top = 50
ot.min_border_bottom = 100
ot.vbar(x='x', top='y', width=0.4, source=source, color='lightgray')
labels = LabelSet(x='x', y='y', text='y', y_offset=10, source=source,
                 text_font_size="11pt", text_align='center')
ot.add_layout(labels)
ot.yaxis.visible = False
ot.xaxis.minor_tick_line_color = None
ot.xgrid.visible = False
ot.ygrid.visible = False
ot.outline_line_color = None

output_notebook()
show(ot)
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, LabelSet
std_sal = {}
filename = 'adult.data'
with open(filename, 'r') as file:
    for dash in file:
        columns = dash.strip().split(',')
        if len(columns) >= 15:
            std = columns[3].strip()
            sal = columns[-1].strip()

            if std not in std_sal:
                std_sal[std] = {"<=50K": 0, ">50K": 0}

            if sal in ["<=50K", ">50K"]:
                std_sal[std][sal] += 1
std_wse = [
    'Preschool', '1st-4th', '5th-6th', '7th-8th', '9th',
    '10th', '11th', '12th', 'HS-grad', 'Some-college',
    'Assoc-acdm', 'Assoc-voc', 'Bachelors', 'Masters',
    'Prof-school', 'Doctorate'
]
std_lvl = [edu for edu in std_wse if edu in std_sal]
less_sal_pct = []
hgh_sal_pct = []

for edu in std_lvl:
    total = std_sal[edu]["<=50K"] + std_sal[edu][">50K"]
    if total > 0:
        less_sal_pct.append((std_sal[edu]["<=50K"] / total) * 100)
        hgh_sal_pct.append((std_sal[edu][">50K"] / total) * 100)
    else:
        less_sal_pct.append(0)
        hgh_sal_pct.append(0)
less_label_y = [val / 2 for val in less_sal_pct]
hgh_label_y = [low + (high / 2) for low, high in zip(less_sal_pct, hgh_sal_pct)]
data = {
    'education': std_lvl,
    'low': less_sal_pct,
    'high': hgh_sal_pct,
    'low_labels': [f"{x:.1f}%" for x in less_sal_pct],
    'high_labels': [f"{x:.1f}%" for x in hgh_sal_pct],
    'low_label_y': less_label_y,
    'high_label_y': hgh_label_y
}

source = ColumnDataSource(data=data)
ot = figure(x_range=std_lvl, height=600, width=1000,
           title="Income Distribution by Education Level")
ot.min_border_bottom = 120

ot.vbar_stack(['low', 'high'], x='education', width=0.6,
             color=['#e74c3c', '#2ecc71'],
             source=source,
             legend_label=['Income ≤50K', 'Income >50K'])
low_labels = LabelSet(x='education', y='low_label_y', text='low_labels',
                      source=source, text_align='center', text_baseline='middle',
                      text_font_size='9pt', text_color='white')
ot.add_layout(low_labels)

high_labels = LabelSet(x='education', y='high_label_y', text='high_labels',
                       source=source, text_align='center', text_baseline='middle',
                       text_font_size='9pt', text_color='white')
ot.add_layout(high_labels)
ot.y_range.start = 0
ot.y_range.end = 100
ot.xaxis.major_label_orientation = 0.8
ot.legend.location = "top_right"
ot.yaxis.visible = False
ot.xaxis.minor_tick_line_color = None
ot.xgrid.visible = False
ot.ygrid.visible = False
ot.outline_line_color = None

show(ot)

output_notebook()
```
